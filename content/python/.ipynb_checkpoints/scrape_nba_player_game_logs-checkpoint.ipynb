```python
def categorise(row):  
    if row['GS'] == 'Inactive':
        return 'no'
    elif row['GS'] == 'Did Not Dress':
        return 'no'
    else:
        return 'yes'

df['Active'] = df.apply(lambda row: categorise(row), axis=1)
df.head()
```

```python
df['Double Digits'] = 0
df.head()
```

```python
df.tail()
```

```python
df = df.reset_index(drop=True)
df.tail()
```

```python
for i in range(len(df)):
    total_double_digits = 0
    if df.loc[i, 'PTS'] >= 10:
        total_double_digits += 1
    if df.loc[i, 'TRB'] >= 10:
        total_double_digits += 1
    if df.loc[i, 'AST'] >= 10:
        total_double_digits += 1
    if df.loc[i, 'STL'] >= 10:
        total_double_digits += 1
    if df.loc[i, 'BLK'] >= 10:
        total_double_digits += 1
    df.loc[i, 'Double Digits'] = total_double_digits
df.head()
```

```python
df['DD Bonus'] = 0
df['TD Bonus'] = 0
df.head()
```

```python
for i in range(len(df)):
    if df.loc[i, 'Double Digits'] == 2:
        df.loc[i, 'DD Bonus'] = 1.5
    elif df.loc[i, 'Double Digits'] >= 3:
        df.loc[i, 'DD Bonus'] = 1.5
        df.loc[i, 'TD Bonus'] = 3
df.head()
```

```python
df['DK Pts'] = 0
```

```python
for i in range(len(df)):
    dk_pts = df.loc[i, 'PTS'] + 0.5*df.loc[i, '3P'] + 1.25*df.loc[i, 'TRB'] + 1.5*df.loc[i, 'AST'] \
            + 2*df.loc[i, 'STL'] + 2*df.loc[i, 'BLK'] - 0.5*df.loc[i, 'TOV'] + df.loc[i, 'DD Bonus'] \
            + df.loc[i, 'TD Bonus']
    df.loc[i, 'DK Pts'] = dk_pts
df.head()
```
